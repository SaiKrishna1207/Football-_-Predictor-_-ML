```python
from numpy import array
from numpy import argmax
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

home = df1['localTeam']
away = df1['visitorTeam']


label_encoder = LabelEncoder()
home_1 = label_encoder.fit_transform(home)
away_1 = label_encoder.fit_transform(away)
#print(integer_encoded)


    
```

```python
print(home_1.shape)
print(away_1.shape)
```

```python
onehot_encoder = OneHotEncoder(categorical_features=[0])
home_1 = home_1.reshape(len(home_1), 1)
away_1 = away_1.reshape(len(away_1), 1)
ohe1 = onehot_encoder.fit_transform(home_1).toarray()
ohe2 = onehot_encoder.fit_transform(away_1).toarray()
#print(onehot_encoded)
```
